```python
import cv2
import os
import pathlib

import numpy as np
```

preprocess video

```python
path = "/Users/j1ah0ng/Desktop/vid.mov"
if not pathlib.Path(path).is_file():
    raise 
```

```python
vc = cv2.VideoCapture(path)
fps = vc.get(5)
n_frames = int(vc.get(7))

w = int(vc.get(3))
h = int(vc.get(4))

frames = np.empty((n_frames, h, w))

i = 0
while vc.isOpened():
    r, f = vc.read()
    if r:
        filt = cv2.Canny(cv2.cvtColor(f, cv2.COLOR_BGR2GRAY), 30, 200)
        frames[i] = filt
        i += 1
        cv2.imshow('', filt)
        key = cv2.waitKey(int(1000 / fps))
        if key == ord('q'):
            break
    else:
        break
vc.release()
cv2.destroyAllWindows()
cv2.waitKey(1)
```
