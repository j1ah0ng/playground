# do I understand Gaussian process regression?

```python
import torch
import torch.nn.functional as F

import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set()

SIGMA = 2
```

set up kernel func

```python
def k(x, y, sigma=SIGMA):
    # radial basis kernel wrt. x, y.
    return torch.exp( -(torch.abs(x - y)**2) / (2 * sigma**2) )
```

```python
k(torch.tensor([1, 2, 3, 4]), 3)
```

set up iteration w/ priors

```python
def Sigma(Xs, sigma=SIGMA):
    # build the covariance matrix guess
    V, H = torch.meshgrid(Xs, Xs, indexing='ij')
    return k(V, H, sigma)
    
```

generate some toy data

```python
n = 100

xs = torch.linspace(-4, 4, n)
ys = F.tanh(xs) + (torch.randn(n) / 8)

sns.lineplot(x=xs, y=ys)
```

let's make some predictions

```python
idxs = torch.randint(0, xs.shape[0], (7,))

xtr = xs[idxs]
ytr = ys[idxs]
Kstar = k(xs[:, None], xtr)

S = Sigma(xtr)
# add back in some variance to avoid zeros and egregious ill-conditioning
S += torch.rand(S.shape) / (SIGMA * 1024)
print(f'k(Sigma) = {torch.linalg.cond(S).item()}')

means = Kstar @ S.inverse() @ ytr
variances = torch.einsum("ij,ij->i", Kstar @ S.inverse().mT, Kstar)
stddevs = variances**0.5
lower = means - 2*stddevs
upper = means + 2*stddevs
```

```python
sns.lineplot(x=xs, y=ys, label='actual', color='b', linewidth=0.8)
sns.scatterplot(x=xtr, y=ytr, label='support', color='b', sizes=(5000,))
sns.lineplot(x=xs, y=means, label='fit', color='r', linewidth=0.5)
sns.lineplot(x=xs, y=lower, label=r'$+2\sigma$', color='r', linewidth=0.5)
c = sns.lineplot(x=xs, y=upper, label=r'$-2\sigma$', color='r', linewidth=0.5)

line = c.get_lines()
plt.fill_between(
    line[2].get_xdata(), 
    line[3].get_ydata(),
    line[2].get_ydata(), 
    color='yellow',
    alpha=0.3,
)
```
